## 1.2. Tipos de dados

### **Dados Categóricos**

*   **Nominais**: Esses dados indicam categorias que não possuem uma ordem ou hierarquia. Exemplo: cores (vermelho, azul, verde), gênero (masculino, feminino)
*   **Ordinais**: Esses dados categóricos têm uma ordem específica ou uma hierarquia. Exemplo: nível de escolaridade (fundamental, médio, superior), avaliação de serviço (ruim, bom, excelente)

### **Dados Numéricos**



*   **Discretos**: Valores numéricos que assumem quantidades contáveis e específicas. Exemplos: número de filhos em uma família, quantidade de carros em um estacionamento
*   **Contínuos**: valores numéricos que podem assumir qualquer valor dentro de um intervalo. Eles incluem números fracionários e infinitos valores possíveis entre quaisquer dois valores. Exemplos: altura de uma pessoa, peso, temperatura, tempo.









## 1.3. Visualização de dados

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
answers = np.array([0, 2, 1, 0, 4, 2, 3, 1, 4, 1, 4, 2, 3])
```

### 1.3.1 Gráfico de Barra
O gráfico de barras é uma representação visual de dados categóricos, muito útil para comparar diferentes categorias ou grupos.

Pode ser usado para uma única amostra (como volume de transações de um ativo) ou para comparação de múltiplas (como volumes de diferentes ativos).

 Ideal para visualizar comparações de volume de negociação entre ações, setores ou classes de ativos. Exemplo: comparar o volume diário de negociação de várias ações.


```python
plt.bar(unique, counts)
plt.xlabel("Respostas")
plt.ylabel("Frequência")
plt.title("Gráfico de Barras das Respostas")
plt.show()
```

## 1.3.2 Histograma
Mostra a distribuição de dados contínuos, agrupando valores em intervalos (ou "bins").

Usado principalmente para uma amostra única, representando a frequência de ocorrência em intervalos.

Útil para visualizar a distribuição de retornos ou variações de preços de um ativo ao longo do tempo, permitindo identificar volatilidade e frequência de valores extremos.

```python
plt.hist(answers, bins=5, edgecolor="black")
plt.xlabel("Respostas")
plt.ylabel("Frequência")
plt.title("Histograma das Respostas")
plt.show()
```

## 1.3.3 Pizza (ou Setores)
Representa a proporção de cada categoria em relação ao total, ideal para visualizar percentuais.

Geralmente usado para uma amostra única, dividindo-a em partes.

Usado para mostrar a composição percentual de um portfólio de investimentos, dividindo os ativos por setor, região, ou tipo de ativo.

```python
plt.pie(counts, labels=unique, autopct="%1.1f%%")
plt.title("Gráfico de Pizza das Respostas")
plt.show()
```

## 1.3.4 Dispersão (Scatter Plot)
Mostra a relação entre duas variáveis numéricas, com cada ponto representando uma observação.

Necessita de duas amostras para mostrar a relação entre elas.

Excelente para analisar a correlação entre o retorno de dois ativos ou entre uma variável econômica (como taxa de juros) e o preço de um ativo.

```python
other_data = np.random.randint(0, 5, len(answers))
plt.scatter(answers, other_data)
plt.xlabel("Respostas")
plt.ylabel("Outro Conjunto")
plt.title("Gráfico de Dispersão")
plt.show()
```

## 1.3.5 Boxplot (Gráfico de Caixa)
Exibe a distribuição de uma amostra com base em quartis, destacando a mediana e possíveis outliers.

Pode ser usado para uma ou múltiplas amostras, sendo muito útil para comparações.

Útil para comparar a dispersão e a mediana dos retornos de diferentes ativos ou portfólios, ajudando a visualizar a volatilidade e identificar outliers.

```python
sns.boxplot(answers)
plt.xlabel("Quantidade")
plt.title("Gráfico Boxplot")
plt.show()
```

## 1.3.6 Linhas
Exibe dados ordenados ao longo do tempo com uma linha conectando os pontos, ideal para séries temporais.

Pode ser usado para uma amostra única ou múltiplas amostras sobrepostas.

Amplamente utilizado para visualizar o preço de fechamento de ativos ao longo do tempo, permitindo identificar tendências e padrões temporais.

```python
plt.plot(answers, marker="o")
plt.xlabel("Índice")
plt.ylabel("Respostas")
plt.title("Gráfico de Linhas das Respostas")
plt.show()
```

## 1.3.7 Área
Semelhante ao gráfico de linhas, mas com a área entre a linha e o eixo preenchida, usado para ver o crescimento ou acúmulo ao longo do tempo.

Geralmente para uma amostra única, mas pode ser usado para múltiplas amostras sobrepostas.

Usado para visualizar o crescimento cumulativo de um índice ou portfólio, mostrando a acumulação de valor ao longo do tempo.

```python
plt.fill_between(range(len(answers)), answers, alpha=0.5)
plt.xlabel("Índice")
plt.ylabel("Respostas")
plt.title("Gráfico de Área das Respostas")
plt.show()
```

```python
sorted_counts = sorted(counts, reverse=True)
cumulative = np.cumsum(sorted_counts) / sum(sorted_counts) * 100

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(range(len(sorted_counts)), sorted_counts, color="C0")
ax2.plot(range(len(cumulative)), cumulative, color="C1", marker="D", ms=5)

ax1.set_xlabel("Categorias")
ax1.set_ylabel("Frequência", color="C0")
ax2.set_ylabel("Frequência Acumulada (%)", color="C1")
plt.title("Diagrama de Pareto")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados simulados para dois ativos
np.random.seed(42)
ativo1 = np.random.normal(0, 1, 100)  # Retornos do Ativo 1
ativo2 = 0.5 * ativo1 + np.random.normal(0, 0.5, 100)  # Retornos do Ativo 2 com alguma correlação com o Ativo 1

# Calculando a correlação
dados = pd.DataFrame({'Ativo 1': ativo1, 'Ativo 2': ativo2})
correlacao = dados.corr()

# Exibindo a correlação e o gráfico de dispersão
print("Matriz de Correlação:\n", correlacao)
sns.heatmap(correlacao, annot=True, cmap="YlGnBu")
plt.title("Mapa de Calor da Correlação entre Ativo 1 e Ativo 2")
plt.show()
```

```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define a list of stock tickers to analyze
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "META"]

# Download the stock data for the past year
data = yf.download(tickers, period="1y")['Close']

# Calculate the correlation matrix for the closing prices
correlation_matrix = data.corr()
'''
é uma medida estatística que descreve a relação entre duas variáveis,
indicando o grau em que elas tendem a variar juntas
Em termos simples, a correlação mostra se as variáveis aumentam ou diminuem juntas e com que intensidade.

'''
# Plotting the heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Heatmap of Stock Prices")
plt.show()
```

## 1.3.10 Diagrama de Densidade
Exibe a distribuição de uma variável contínua como uma curva suave, estimando a função densidade de probabilidade.

Usado para uma amostra única, visualizando a densidade.

Usado para visualizar a distribuição dos retornos diários de um ativo, ajudando a entender o comportamento e a probabilidade de diferentes retornos.


```python
# Calculate the daily returns
daily_returns = data.pct_change().dropna()

# Probability and Density Diagram for Daily Returns
plt.figure(figsize=(12, 6))

# Plot KDE for each stock's daily returns
for ticker in tickers:
    sns.kdeplot(daily_returns[ticker], label=ticker, fill=True, linewidth=2)

# Improving the presentation
plt.xlabel("Daily Return")
plt.ylabel("Density")
plt.title("Density Plot of Daily Returns for Selected Stocks")
plt.legend(title="Stocks")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# 2.2.2 Desvio Padrão
O desvio padrão é a raiz quadrada da variância, o que o coloca na mesma unidade dos dados originais.

Ele mede o quão "espalhados" os dados estão em torno da média. Por ser a raiz quadrada da variância, ele é mais intuitivo e fácil de interpretar.

No mercado financeiro, o desvio padrão e a variância dos retornos de um ativo são usados para medir o risco.

Quanto maior o desvio padrão dos retornos de um ativo, maior o risco (volatilidade) associado a ele

## 2.3. Distribuições de probabilidade

A distribuição de probabilidade descreve todos os valores possíveis e a probabilidade de ocorrência de cada um deles em um conjunto de dados ou evento aleatório

é uma função que atribui uma probabilidade para cada possível resultado de um experimento aleatório, ajudando a entender o comportamento esperado dos dados.

Por exemplo, em um contexto de retornos financeiros, uma distribuição de probabilidade pode mostrar como os retornos de um ativo variam ao longo do tempo e quais retornos são mais ou menos prováveis de ocorrer.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parâmetros para a distribuição normal
media = 0  # Média dos retornos
desvio_padrao = 0.02  # Desvio padrão dos retornos
tamanho_amostra = 1000  # Número de retornos simulados

# Gerando uma amostra de retornos simulados seguindo uma distribuição normal
retornos_simulados = np.random.normal(media, desvio_padrao, tamanho_amostra)

# Plotando a distribuição dos retornos simulados
plt.figure(figsize=(10, 6))
sns.histplot(retornos_simulados, kde=True, bins=30, color="skyblue")
plt.xlabel("Retorno Simulado")
plt.ylabel("Frequência")
plt.title("Distribuição de Probabilidade dos Retornos Simulados")
plt.grid(True)
plt.show()


'''
Interpretação
A curva resultante é simétrica e concentrada em torno da média (0%), com a maioria dos retornos caindo próximo da média e poucos valores em áreas mais afastadas.
No mercado financeiro, isso ajudaria um investidor a entender que a maior parte dos retornos diários estará em torno de uma média, com valores extremos sendo mais raros.
Essa informação é crucial para avaliar a expectativa de retorno e o risco envolvido.

'''
```

```python
## Distribuição da NORMAL Da APPL
# media_appl
# apple_dp
mean = data['AAPL'].mean()
tamanho_amostra = len(data['AAPL'])
retornos_apple = data['AAPL'].pct_change().dropna()
retornos = np.random.normal(mean, apple_dp, tamanho_amostra)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(retornos, kde=True, bins=30, color="skyblue")
plt.xlabel("Retorno APPL")
plt.ylabel("Frequência")
plt.title("Distribuição de Probabilidade APPL")
plt.grid(True)
plt.show()
```
